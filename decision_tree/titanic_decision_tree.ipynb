```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/titanic.csv')
df
```

```python
df = df.drop(['PassengerId', 'Name', 'Ticket','SibSp', 'Cabin','Embarked','Parch'], axis=1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df
```

```python
input_n = df.drop('Survived', axis=1)
target = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_n, target, test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(15,10))
tree.plot_tree(model, filled=True)
plt.legend()
plt.show()
```

```python
model.score(x_test, y_test)
```
