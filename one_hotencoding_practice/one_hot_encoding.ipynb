The file  carprices.csv has car sell prices for 3 different models. First plot data points on a scatter plot chart to see if linear regression model can be applied. If yes, then build a model that can answer following questions,

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv("/content/drive/MyDrive/Colab Notebooks/carprices.csv")
df
```

```python

dummies = pd.get_dummies(df['Car Model'], drop_first=True)
dummies


df_with_dummies = pd.concat([df, dummies], axis='columns')

df_with_dummies
```

```python
df_with_dummies.drop('Car Model', axis=1, inplace=True)
df_with_dummies
```

```python
df_independent = df_with_dummies.drop('Sell Price($)', axis=1)
df_independent
```

```python
df_dependent = df_with_dummies['Sell Price($)']
df_dependent
```

```python
reg = LinearRegression()
reg.fit(df_independent, df_dependent)
```

1) Predict price of a mercedez benz that is 4 yr old with mileage 45000

```python
reg.predict([[45000,4,False,True]])
```

2) Predict price of a BMW X5 that is 7 yr old with mileage 86000

```python
reg.predict([[86000,7,True,False]])
```

3) Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

```python
 reg.score(df_independent,df_dependent)
```
